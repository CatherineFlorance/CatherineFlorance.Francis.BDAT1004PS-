## Write a class named Marsupial that can be used as shown below:Â¶

```python
class Marsupial:
    def __init__(self):
        self.pouch = [] ##list data structure that stores items contained in the list

    def put_in_pouch(self, item):
        self.pouch.append(item) ## appends items to the self pouch

    def pouch_contents(self):
        return self.pouch ## returns the contents in the self pouch

m = Marsupial()
m.put_in_pouch('doll')
m.put_in_pouch('firetruck')
m.put_in_pouch('kitten')
print(m.pouch_contents())


class Kangaroo(Marsupial):
    def __init__(self, x, y):
        # Call the parent class (Marsupial) constructor to initialize the pouch
        super().__init__()

        # Initialize the coordinates of the Kangaroo
        self.x = x
        self.y = y

    def jump(self, dx, dy):
        # Update the Kangaroo's coordinates by adding the displacement values
        self.x += dx
        self.y += dy

    def __str__(self):
        # Return a string representation of the Kangaroo's location
        return f"I am a Kangaroo located at coordinates ({self.x},{self.y})"
    
# Create an instance of Kangaroo with initial coordinates (0,0)
k = Kangaroo(0, 0)

# Print the string representation of the Kangaroo
print(k)

# Add items to the Kangaroo's pouch
k.put_in_pouch('doll')
k.put_in_pouch('firetruck')
k.put_in_pouch('kitten')

# Print the Kangaroo's pouch contents
print(k.pouch_contents())

# Make the Kangaroo jump multiple times
k.jump(1, 0)
k.jump(1, 0)
k.jump(1, 0)

# Print the updated string representation of the Kangaroo
print(k)

```

### Write a recursive method binary() that takes a non-negative integer n and prints the binary representation of integer n.

```python
def binary(n):
    if n > 1:
        binary(n // 2)
    print(n % 2, end='')
binary(0)
binary(1)
binary(3)
binary(9)
```

### Implement a class named HeadingParser that can be used to parse an HTML document, and retrieve and print all the headings in the document. You should implement your class as a subclass of HTMLParser, defined in Standard Library module html.parser. When fed a string containing HTML code, your class should print the headings, one per line and in the order in which they appear in the document. Each heading should be indented as follows: an h1 heading should have indentation 0, and h2 heading should have indentation 1, etc. Test your implementation using w3c.html

```python
from html.parser import HTMLParser


class HeadingParser(HTMLParser):
    def __init__(self):
        super().__init__()
        self.headings = []
        self.current_heading = ''

    def handle_starttag(self, tag, attrs):
        if tag.startswith('h') and len(tag) == 2 and tag[1].isdigit():
            self.current_heading = int(tag[1]) - 1

    def handle_endtag(self, tag):
        if tag.startswith('h') and len(tag) == 2 and tag[1].isdigit():
            self.current_heading = ''

    def handle_data(self, data):
        if self.current_heading != '':
            indentation = '  ' * self.current_heading
            heading_text = data.strip()
            self.headings.append(f'{indentation}{heading_text}')

    def print_headings(self):
        for heading in self.headings:
            print(heading)


# Example usage
with open('W3C.html') as infile:
    content = infile.read()

hp = HeadingParser()
hp.feed(content)
hp.print_headings()
```

### Implement recursive function webdir() that takes as input: a URL (as a string) and non-negative integers depth and indent. Your function should visit every web page reachable from the starting URL web page in depth clicks or less, and print each web page's URL. As shown below, indentation, specified by indent, should be used to indicate the depth of a URL.

```python
import requests
from bs4 import BeautifulSoup


def webdir(url, depth, indent):
    if depth < 0:
        return

    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    print(' ' * indent + url)

    if depth == 0:
        return

    for link in soup.find_all('a'):
        href = link.get('href')
        if href and href.startswith('http'):
            webdir(href, depth - 1, indent + 1)


# Example usage
webdir('http://reed.cs.depaul.edu/lperkovic/csc242/test1.html', 2, 0)
```
